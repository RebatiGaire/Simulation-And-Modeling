```
# Import Libraries

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
# Initialization


G = np.array([27210,36333,37902,36307,32567,34393,29630,38149,27590,37048,
              27486,31001,35886,38824,25541,41658,38738,33249,32281,32488,
              29820,33661,33704,39534,39101,36242,31193,39420,34123,30663,
              41841,40642,34452,35827,35153,27947,29723,32947,28379,40041,
              27700,28292,27243,28325,32278,29910,41544,32173,27511,41192,
              ])

N = len(G)  # No of year
C = np.zeros(shape= (N))
I = np.zeros(shape= (N))
T = np.zeros(shape= (N))
Y = np.zeros(shape= (N))

Y[-1] = 150000
```

```
for i in range(0, N):
  
  I[i] = 2 + 0.1 * Y[i-1] # present Investment amount from previous year's national income
  T[i] = 0.2 * Y[i-1] # Present Tax amount from previous year's national income
  C[i] = 20 + 0.7 * (Y[i-1] - T[i]) # present Consumed amount (diff of previous year's income and present year's tax)
  Y[i] = 45.45 + 2.27 * (I[i] + G[i]) # This year's national income

```

```
x = range(N)
plt.figure(figsize= (12, 8))
plt.plot(x, I, 'r', label= 'Investment')
plt.plot(x, C, 'b', label= 'Consumption')
plt.plot(x, Y, 'y', label= 'National Income')
plt.plot(x, T, 'g', label= 'Taxes')
plt.plot(x, G, label= 'Government Expenditure')
plt.legend()
plt.title('National Econometric System')
plt.xlabel('Time in Years')
plt.ylabel('Amount of Money in Million Rs.')
plt.show()
```
